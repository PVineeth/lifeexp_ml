```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
df = pd.read_pickle("../../Data/New/Main_Data/Dataset_Removed_Countries.pkl")
```

```python
xxxx = df.columns[df.isnull().any()]
xxxx
```

```python
for country in df['Country'].unique():
    for column in range(0, len(xxxx)):
        print(column)
        if(df[df['Country'] == country][xxxx[column]].isnull().any()):
            print(xxxx[column])
            print(country, column)
            # DEALING WITH MISSING VALUES
            imp = SimpleImputer(strategy='mean')
            replaced_array = imp.fit_transform(df[df['Country'] == country][[xxxx[column]]])
            print(replaced_array)
            # Replacing Values
            count = 0
            start_index = df[df['Country'] == country][xxxx[column]].index[0]
            end_index = df[df['Country'] == country][xxxx[column]].index[-1]
            for row in range(start_index, end_index+1):
                df.at[row, xxxx[column]] = replaced_array[count][0]
                count+=1
```

```python
df.Life_Expectancy = df.Life_Expectancy.astype(np.float64)
df.Adult_Mortality = df.Adult_Mortality.astype(np.int64)
df.Infant_Deaths = df.Infant_Deaths.astype(np.float64)
df.Alcohol = df.Alcohol.astype(np.float64)
df.Percentage_Expenditure = df.Percentage_Expenditure.astype(np.float64)
df.Measles = df.Measles.astype(np.float64)
df['BMI_5-19_years']  = df['BMI_5-19_years'] .astype(np.float64)
df['BMI_18+_years']  = df['BMI_18+_years'].astype(np.float64)
df.Polio= df.Polio.astype(np.int64)
df['Under-Five_Deaths']= df['Under-Five_Deaths'].astype(np.float64)
df.Total_Expenditure = df.Total_Expenditure .astype(np.float64)
df.Diphtheria = df.Diphtheria .astype(np.int64)
df.GDP = df.GDP.astype(np.float64)
df['Thinness_10-19_years']  = df['Thinness_10-19_years'].astype(np.float64)
df['Thinness_5-9_years']   = df['Thinness_5-9_years'].astype(np.float64)
df.Income_Composition_Of_Resources = df.Income_Composition_Of_Resources.astype(np.float64)
df.Schooling  = df.Schooling.astype(np.float64)
```
